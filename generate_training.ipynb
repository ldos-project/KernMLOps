```python
from python.kernmlops.analysis.collector import Collector
```

```python
w = open("hello.txt", "bw")
collect = Collector("./config/redis_never.yaml")
collect.start_collection(logfile=w)
redis_never_info_inner = collect.wait()
print("Never Finished")
collect = Collector("./config/redis_always.yaml")
collect.start_collection(logfile=w)
redis_always_info_inner = collect.wait()
print("Always Finished")
```

```python

redis_never_info = [redis_never_info_inner]
redis_always_info = [redis_always_info_inner]
for i in range(29):
    w = open(f"hello_{i}.txt", "bw")
    collect = Collector("./config/redis_never.yaml")
    collect.start_collection(logfile=w)
    redis_never_info_inner = collect.wait()
    redis_never_info.append(redis_never_info_inner)
    print("Never Finished")
    collect = Collector("./config/redis_always.yaml")
    collect.start_collection(logfile=w)
    redis_always_info_inner = collect.wait()
    redis_always_info.append(redis_always_info_inner)
    print("Always Finished")
    close(w)
```
