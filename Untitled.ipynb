```python
import numpy as np
import matplotlib.pyplot as plt

def generate_uniform_sum(n, total_sum):
    cuts = np.sort(np.random.uniform(0, total_sum, n-1))
    parts = np.diff(np.concatenate(([0], cuts, [total_sum])))
    return parts

# Generate a large number of sets of 4 numbers that sum to 100
samples = np.array([generate_uniform_sum(4, 100) for _ in range(10000)])

# Plot histograms of the parts
plt.figure(figsize=(12, 8))

for i in range(4):
    plt.subplot(2, 2, i+1)
    plt.hist(samples[:, i], bins=30, density=True)
    plt.title(f"Distribution of x{i+1}")
    plt.xlabel(f"x{i+1}")
    plt.ylabel("Density")

plt.tight_layout()
plt.show()
```
